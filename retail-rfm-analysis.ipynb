# Business Problem

Today's consumer behaviors significantly vary, generating diverse revenue streams for companies. This variability in shopping habits necessitates that companies segment their customers based on their behaviors, compelling them to develop unique strategies for each segment. Additionally, predicting the potential loyalty of customers to the company is of utmost importance. Every large corporation must conduct detailed analyses of customer behavior and act accordingly based on these analyses.

FLO, recognized as one of the leading footwear retailers in Turkey, aims in this project to segment its customer base using RFM (Recency, Frequency, Monetary) Analysis. This approach is intended to more effectively understand and meet the varying needs of its customers.


# The Story of Dataset
This dataset captures the shopping activities of customers who engaged in purchasing from Flo through both online and offline channels (OmniChannel) during 2020-2021. It includes:

- **master_id:** A distinct identifier for each customer.
- **order_channel:** The shopping platform utilized by the customer (Android, iOS, Desktop, Mobile).
- **last_order_channel:** The platform on which the customer's most recent transaction occurred.
- **first_order_date:** The date when the customer made their inaugural purchase.
- **last_order_date:** The date of the customer's most recent purchase.
- **last_order_date_online:** The date on which the customer last made an online purchase.
- **last_order_date_offline:** The date on which the customer last made an offline purchase.
- **order_num_total_ever_online:** The aggregate number of online purchases executed by the customer.
- **order_num_total_ever_offline:** The aggregate number of offline purchases executed by the customer.
- **customer_value_total_ever_online:** The cumulative amount spent by the customer on online purchases.
- **customer_value_total_ever_offline:** The cumulative amount spent by the customer on offline purchases.
- **interested_in_categories_12:** The categories in which the customer has made purchases over the past 12 months.
- **interested_in_categories_12:** List of categories the customer has shopped in the last 12 months

# Calculating RFM Metrics


```python
df["last_order_date"].max()
```

```python
# Present Date (date of analysis)

today_date = dt.datetime(2021, 6, 1)
```

```python
today_date
```

```python
# Calculating RFM Metrics

rfm = df.groupby("master_id").agg({"last_order_date":lambda last_order_date:(today_date-last_order_date.max()),
                                   "Omnichannel": lambda Omnichannel : Omnichannel.sum(),
                                   "TotalPrice":lambda TotalPrice : TotalPrice.sum()})
```

```python
rfm.head()
```

```python
# Changing column names

rfm.columns = ['recency', 'frequency', 'monetary']
```

```python
rfm.head()
```

# Calculating RFM Scores (Recency, Frequency, Monetary)


```python
# Calculating Recency_Score

rfm["recency_score"] = pd.qcut(rfm['recency'], 5, labels=[5, 4, 3, 2, 1])
```

```python
rfm.head()
```

```python
# Calculating Frequency_Score

rfm["frequency_score"] = pd.qcut(rfm['frequency'].rank(method="first"), 5, labels=[1, 2, 3, 4, 5])
```

```python
rfm.head()
```

```python
# Calculating Monetary_Score

rfm["monetary_score"] = pd.qcut(rfm['monetary'], 5, labels=[1, 2, 3, 4, 5])
```

```python
rfm.head()
```

```python
# Creating RFM_SCORE VARIABLE

rfm["RFM_SCORE"] = (rfm['recency_score'].astype(str) + rfm['frequency_score'].astype(str))
```

# Creating & Analysing RFM Segments


![](https://miro.medium.com/max/617/1*oUKMc3gk53hccFX4YNeu1g.png)

* **Champions** are the best customers, who bought most recently, most often, and are heavy spenders. Reward these customers. They can be the first to buy new products

* **Potential Loyalists** are the recent customers with average frequency and who spent a good amount. You can make them your Loyalists or Champions by offering membership or loyalty programs or recommending related products to them.

* **New Customers** are the customers who have a high overall RFM score but are not frequent shoppers. Special offers can be provided to increase their visits

* **At Risk Customers** are the customers who purchased often and spent big amounts, but havent purchased recently. Personalized deals and product/service recommendations can help reactivate them

* **Cant Lose Them** are the customers who used to visit and purchase quite often, but havent been visiting recently. Get them to revisit with relevant promotions and conduct surveys to find out what went wrong and not lose them to a competitor.



```python
# Creation of a segment map
seg_map = {
    r'[1-2][1-2]': 'hibernating',
    r'[1-2][3-4]': 'at_Risk',
    r'[1-2]5': 'cant_loose',
    r'3[1-2]': 'about_to_sleep',
    r'33': 'need_attention',
    r'[3-4][4-5]': 'loyal_customers',
    r'41': 'promising',
    r'51': 'new_customers',
    r'[4-5][2-3]': 'potential_loyalists',
    r'5[4-5]': 'champions'
}
```

```python
# Creating segment variable

rfm['segment'] = rfm['RFM_SCORE'].replace(seg_map, regex=True)
```

```python
rfm = rfm.reset_index()
```
